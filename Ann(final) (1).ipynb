```python
# 3. Prepare X and y
X = data.iloc[:, 3:-1]
y = data.iloc[:, -1]
```

```python
# 4. Encode categorical columns
le = LabelEncoder()
X['Gender'] = le.fit_transform(X['Gender'])  # Male/Female to 0/1
X = pd.get_dummies(X, columns=['Geography'], drop_first=True)  # One-Hot Encode Geography
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# 6. Scale features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# 7. Build ANN
ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
# 8. Compile ANN
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# 9. Train ANN
history = ann.fit(X_train, y_train, batch_size=32, epochs=50, validation_data=(X_test, y_test))
```

```python
# 10. Predict and Evaluate
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# 11. Evaluate
print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
